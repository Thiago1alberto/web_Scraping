```python
import requests
from bs4 import BeautifulSoup

def buscar_dados_ca(numero_ca):
    """
    Requisição ao site ConsultaCA para extrair os dados do EPI correspondente ao CA informado.
    """
    url = f"https://consultaca.com/{numero_ca}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        try:
            titulo = soup.find("h1").get_text(strip=True)
            validade_element = soup.find("span", class_="validade_ca")
            validade_ca = validade_element.get_text(strip=True) if validade_element else "Não encontrado"
            return {"CA": numero_ca, "Nome do EPI": titulo, "Validade do CA": validade_ca}
        except AttributeError:
            return None  # Ignora CAs inválidos
    return None
```

```python
import time
from openpyxl import load_workbook
from pandas import ExcelWriter

intervalo_inicial = 0
intervalo_final = 60000  # Pode ser ajustado

for ca in range(intervalo_inicial, intervalo_final + 1):
    resultado = buscar_dados_ca(ca)
    if resultado:
        print(f"Encontrado: {resultado}")
        df_temp = pd.DataFrame([resultado])
        with ExcelWriter(caminho_arquivo, mode="a", if_sheet_exists="overlay", engine="openpyxl") as writer:
            df_temp.to_excel(writer, index=False, header=False, startrow=writer.sheets['Sheet1'].max_row)
    time.sleep(0.2)  # Delay para evitar bloqueio do site
```
